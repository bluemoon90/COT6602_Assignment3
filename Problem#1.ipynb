### Solution

As in the proof of theorem 2 (Quantum NP - A Survey by Dorit Aharonov and Tomer Naveh), the verifier needs to ask the prover to provide him with polynomially many copies of the witness. 

We construct a new verifier which runs in parallel polynomially many copies of the verifier V, then outputs the majority. 

We know that the probability that the first copy of V outputs 1 is less than 1/3. After the first verifier was applied, we can apply the second verifier. The second verifier gets as an input some state, which can be conditioned on the result of the measurement of the first verifier. However, regardless of what this output was, it is still correct that the probability for an output 1 is less than 1/3. 

Hence, the probability for the majority of the verifiers being 1 can be bounded from above by the probability that polynomially many independent Bernoulli trials with bias 1/3 will be 1, which decays exponentially.



```python
a = [1/3]
for i in range(10):
    a.append(a[-1]/3)

plt.figure(figsize=(8, 6))
plt.plot(a)
plt.xlabel("$n$")
plt.ylabel("uper bound for probability")
plt.show()
```
